```python
import xml.etree.ElementTree as ET
import os
import json

#used to build a dictionary of element tag and direct children
#recursion stops when it gets to end of file. meaning no more tags left
# for each tags it iterates over the children and updats the schema
def build_schema_tree(node, schema):
    tag = getattr(node, 'tag', None)
    if tag is None:
        return
    if tag not in schema:
        schema[tag] = set()
    for child in list(node):
        schema[tag].add(child.tag)
        build_schema_tree(child, schema)
    print(schema)

#XML parses using etree
# starting from root element it recursively fills the schema with parent-child mapping
def XML_parser(path):
    tree = ET.parse(path)
    root = tree.getroot()
    schema = {}
    build_schema_tree(root, schema)
    result = {}
    for k in schema:
        result[k] = list(schema[k])
    return result
```

```python
MS ='mediated.xml'

schema = XML_parser(MS)
print("MS schema tree")
print(json.dumps(schema, indent=2))
with open("Mediated_schema.json", "w") as f:
    json.dump(schema, f, indent=2)
```

```python
DS ='movie.xml'

schema = XML_parser(DS)
print("DS schema tree")
print(json.dumps(schema, indent=2))
with open("Data_souce_schema.json", "w") as f:
    json.dump(schema, f, indent=2)
       
```
